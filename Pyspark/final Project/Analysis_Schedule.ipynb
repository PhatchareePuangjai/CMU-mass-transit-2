```python
from pyspark.sql import functions as F

def countNumberStudent(day, semester, building):
    
    data = df_s.filter( df_s.day.like(day) & df_s.building.like(building) & df_s.semester.like(semester)).withColumn('student_regist', df_s.student_regist)
    
    total = (data.groupBy().agg(F.sum("student_regist")).collect())[0]
    
    return (data,total)
```

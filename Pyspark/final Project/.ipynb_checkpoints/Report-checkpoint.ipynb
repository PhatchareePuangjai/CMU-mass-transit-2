```python
# เลือกเส้นทาง
df_1 = df.filter( df.route.like('1') )
df_3 = df.filter( df.route.like('3') )

# ตัดข้อมูลที่  speed = 0
df_1 = df_1.filter(df_1.speed != 0)
df_3 = df_3.filter(df_3.speed != 0)
```

```python
from pyspark.sql import functions as F

df_3_day = df_3.select('bus', 'driver', 'speed' , 'timestamp' , F.date_format('timestamp','yyyy-MM-dd').alias('day')).groupby('day', 'bus', 'driver').agg(F.avg("speed")).sort('day')
df_1_day = df_1.select('bus', 'driver', 'speed' , 'timestamp' , F.date_format('timestamp','yyyy-MM-dd').alias('day')).groupby('day', 'bus', 'driver').agg(F.avg("speed")).sort('day')
```

```python
df_3_day.show()
```

```python
arr_speed = []
keep_data = []

#นับจำนวน df_3_day
len_3 = (df_3_day.count())
len_1 = (df_1_day.count())

#แปร เป็น array
df_3_day = df_3_day.collect()
df_1_day = df_1_day.collect()
```

```python
for i in range(len_3):
    keep_data = []
    day_ = df_3_day[i][0]
    keep_day = (day_).split('-')
    year = str( int(keep_day[0]) + 543 )
    day = year + '-' + keep_day[1] + '-' + keep_day[2]
    
    keep_data.append(day)
    keep_data.append(df_3_day[i][1])
    keep_data.append(df_3_day[i][2])
    keep_data.append('3')
    keep_data.append(df_3_day[i][3])
    
    arr_speed.append(keep_data)
    
# print(arr_speed)
```

```python
# create dataframe pyspark
from pyspark.sql.types import *

cSchema = StructType([ StructField("day", StringType())\
                      ,StructField("bus", IntegerType())\
                      ,StructField("driver", StringType())\
                      ,StructField("speed", DoubleType())\
                     ])
```

```python
df_newData = spark.createDataFrame(arr_speed,schema=cSchema) 
```

```python
df_newData.show()
df_newData.printSchema()
```

```python
# เขียนลง database
# --------------------------------------------------------------------------------------------------------------------------

df_newData.write.format("com.mongodb.spark.sql.DefaultSource").mode("append")\
    .option("uri","mongodb://cmuuser:cmuuserp%40ssw0rd@27.254.90.146:27017/timesyscmu.transit_Speed?authSource=timesyscmu")\
    .save()

# --------------------------------------------------------------------------------------------------------------------------
```

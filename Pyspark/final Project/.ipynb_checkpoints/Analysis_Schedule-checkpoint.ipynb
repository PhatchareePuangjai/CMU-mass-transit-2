```python
# ====================== ให้ numpy และ pandas ====================== 
# ====================== เเต่ไม่ใช่งานจริง ======================

import pandas as pd
import numpy as np
import pymongo
from pymongo import MongoClient
     
def countNumberStudent(data, semester, building):
    
    num_day = [0,0,0,0,0,0,0]
    len_data = len(data)
#     len_data = 50
    for i in range (len_data):
        day = str(data['day'][i])
        

        # ตึกเรียน และเลือกปีการศึกษา
        if(data['semester'][i] == semester and data['building'][i] == building):
            # แยกวิชาที่เรียน 2 วัน    
            if day.find(",") != -1:
                day2 = day.split(",")
                
                if(day2[0] == "Monday" or day2[1] == "Monday"):
                    num_day[0] = num_day[0] + int(data['student_regist'][i])
                    
                if(day2[0] == "Tuesday" or day2[1] == "Tuesday"):
                    num_day[1] = num_day[1] + int(data['student_regist'][i])
                    
                if(day2[0] == "Wednesday" or day2[1] == "Wednesday"):
                    num_day[2] = num_day[2] + int(data['student_regist'][i])
                    
                if(day2[0] == "Thursday" or day2[1] == "Thursday"):
                    num_day[3] = num_day[3] + int(data['student_regist'][i])
                    
                if(day2[0] == "Friday" or day2[1] == "Friday"):
                    num_day[4] = num_day[4] + int(data['student_regist'][i])
                    
                if(day2[0] == "Saturday" or day2[1] == "Saturday"):
                    num_day[5] = num_day[5] + int(data['student_regist'][i])
                    
                if(day2[0] == "Sunday" or day2[1] == "Sunday"):
                    num_day[6] = num_day[6] + int(data['student_regist'][i])

            # วิชาที่เรียน วันเดียว
            else:
                
                if(day == "Monday"):
                    num_day[0] = num_day[0] + int(data['student_regist'][i])
                elif(day == "Tuesday"):
                    num_day[1] = num_day[1] + int(data['student_regist'][i])
                elif(day == "Wednesday"):
                    num_day[2] = num_day[2] + int(data['student_regist'][i])
                elif(day == "Thursday"):
                    num_day[3] = num_day[3] + int(data['student_regist'][i])
                elif(day == "Friday"):
                    num_day[4] = num_day[4] + int(data['student_regist'][i])
                elif(day == "Saturday"):
                    num_day[5] = num_day[5] + int(data['student_regist'][i])
                elif(day == "Sunday"):
                    num_day[6] = num_day[6] + int(data['student_regist'][i])
#             print(num_day)
#             print(day)
#             print(int(data['student_regist'][i]))
#             print()
              
    return num_day

def main():
    
    #point the client at mongo URI
    client = MongoClient("mongodb://cmuuser:cmuuserp%40ssw0rd@27.254.90.146:27017/timesyscmu.classSchedule?authSource=timesyscmu")
    #select database
    db = client['timesyscmu']
    #select the collection within the database
    data = db.classSchedule
    #convert entire collection to Pandas dataframe
    data = pd.DataFrame(list(data.find()))
    
    num_day_PS_1 = countNumberStudent(data, "1", "Political Science")
    num_day_PS_2 = countNumberStudent(data, "2", "Political Science")
    num_day_CS_1 = countNumberStudent(data, "1", "Computer Science")
    num_day_CS_2 = countNumberStudent(data, "2", "Computer Science")
    
    print(num_day_PS_1)
    print(num_day_PS_2)
    print(num_day_CS_1)
    print(num_day_CS_2)
    
#     หาช่วงเวลาทั้งหมด
    
    
#     data3 = data.loc[data['year'] == "2561"]

    grouped = data.groupby('time_start').sum()[['student_regist']]
   
        
if __name__ == '__main__':
    main()

                
```

```python
from pyspark.sql import functions as F

def countNumberStudent(day, semester, building):
    
    data = df_s.filter( df_s.day.like(day) & df_s.building.like(building) & df_s.semester.like(semester)).withColumn('student_regist', df_s.student_regist)
    
    total = (data.groupBy().agg(F.sum("student_regist")).collect())[0]
    
    return (data,total)

day_Mo,total_Mo = countNumberStudent("%Monday%", "1", "Political Science")
day_Tu,total_Tu = countNumberStudent("%Tuesday%", "1", "Political Science")
day_We,total_We = countNumberStudent("%Wednesday%", "1", "Political Science")
day_Th,total_Th = countNumberStudent("%Thursday%", "1", "Political Science")
day_Fr,total_Fr = countNumberStudent("%Friday%", "1", "Political Science")
day_Sa,total_Sa = countNumberStudent("%Saturday%", "1", "Political Science")
day_Su,total_Su = countNumberStudent("%Sunday%", "1" , "Political Science")


day_Mo.show()
day_Tu.show()
day_We.show()
day_Th.show()
day_Fr.show()
day_Sa.show()
day_Su.show()


# แสดงนักจำนวนนักศึกษาทั้งหมดที่มีเรียนในตึก
print ("จำนวนนักศึกษาวันจันทร์: %d" %(total_Mo[0])) 
print ("จำนวนนักศึกษาวันอังคาร: %d" %(total_Tu[0])) 
print ("จำนวนนักศึกษาวันอังคาร: %d" %(total_We[0])) 
print ("จำนวนนักศึกษาวันพฤหัสบดี: %d" %(total_Th[0])) 
print ("จำนวนนักศึกษาวันศุกร์: %d" %(total_Fr[0])) 
print ("จำนวนนักศึกษาวันเสาร์: %d" %(total_Sa[0])) 
print ("จำนวนนักศึกษาวันอาทิตย์: %d" %(total_Su[0])) 

# ---------------------------------------------------------------------
# print the list

# list_elements = day_Mo.collect()
# for element in list_elements:
#     print(element.numStudents)
```

```python
# จัดรูปแบบข้อมูลเพื่อ จะนำลงไปเก็บใน database

def getData(day, time, data, num_all, num_new, num_end):
    time_all = ['08.00','08.30','09.00','09.30','10.00','10.30','11.00','11.30','12.00','12.30','13.00','13.30','14.00','14.30','15.00','15.30','16.00','16.30','17.00']
    
    for x in time.collect():
            keep = []

            # x[0] = time , x[1] = numStudent
            # แทนที่ . ในช่วงเวลาด้วย :
            time2 = x[0].split('-')
            
            for i in range (len(time_all)):
                
                if(float(time2[0]) <= float(time_all[i]) and float(time_all[i]) <= float(time2[1])):
                    
                    num_all[i] = num_all[i] + int(x[1])
                    
                    # เริ่มเก็บจำนวนที่เริ่มเรียน
                    if(float(time2[0]) == float(time_all[i])):
                        num_new[i] = num_new[i] + int(x[1])
                        
                    # เก็บจำนวนที่เลิกเรียน
                    if(float(time_all[i]) == float(time2[1])):
                        num_end[i] = num_end[i] + int(x[1])
        
    return (num_all,num_new,num_end)
```

```python
from copy import deepcopy

from time import time
from datetime import datetime
from dateutil.parser import parse
import geopy.distance


def distances(coords_1,coords_2):
    
    return (geopy.distance.vincenty(coords_1, coords_2).km)

def cut_day(time_Mo):
    
    day_Mo = []
    for i in time_Mo:
        keep = []

        for x in df_Mo.collect():
#             print(i)
            time = (x.timestamp).split('T')

            if(time[0] == i):
                keep.append(x)

        day_Mo.append(keep)
        
    return day_Mo
#========================================

def diff_times_in_seconds(t1, t2):
    # caveat emptor - assumes t1 & t2 are python times, on the same day and t2 is after t1
    h1, m1, s1 = t1.hour, t1.minute, t1.second
    h2, m2, s2 = t2.hour, t2.minute, t2.second
    
    #print(t1, h1, m1, s1)
    #print(t2, h2, m2, s2)
    
    t1_secs = s1 + 60 * (m1 + 60*h1)
    t2_secs = s2 + 60 * (m2 + 60*h2)
    t3_min = (t2_secs - t1_secs)/60
    
    return (t3_min)

#========================================

def averageTime(day1):

    Avg = 0
    keep = 0
    count = 0
    cur = []                 #เอาไว้วน bus number
    keep_cur = []
    Avg_all = []
    
    time_star = [7.00, 8.01, 9.31, 11.01, 12.31, 13.01, 14.31, 16.01, 17.31, 19.01, 20.31]
    time_end =  [8.00, 9.30, 11.00, 12.30, 13.00, 14.30, 16.00, 17.30, 19.00, 20.30, 22.00]

    for i in range(11):

        # set value start
        Avg = 0
        keep = 0
        count = 0
        cur = []
        keep_cur = []

        # เลือกช่วงเวลา
        time_s = time_star[i]
        time_e = time_end[i]

        for x in day1:

            #หาระยะห่างจากจุด start ที่เริ่มปล่อยรถ
            dis_start = (distances( (18.7994182212734, 98.952321875), (x.lat, x.lng) ))     

            #หาระยะห่างจากจุดหน้าตึกเรียน 
            #  หน้ามอ 18.8076975,98.9546958
            # อ.มช. 18.8002555,98.9533655
            
            
            # ภาคคอม 18.80402913353402, 98.9527465659744
            # รัฐศาสตร์ 18.8049546,98.9543871
#             print(lat,lng)
            dis_destination = (distances( (18.8049546,98.9543871), (x.lat, x.lng) ))   

            # ตัดเวลาเเละทำเวลาเป็น float
            time = (x.timestamp).split('T')

            #time_float เอาไว้เลือกช่วงเวลา , time_now เอาไว้คำนวนระยะห่างของเวลา
            time_float = time[1].split(':')
            time_float = float(time_float[0] + '.' + time_float[1])
            
            time1 = datetime.strptime((x.timestamp), '%Y-%m-%dT%H:%M:%S.%f')
            time_now = (time1.time())


            #-------------------------------------------------------------------------------
            #แยกเวลาเปนแต่ละช่วงเวลา    

            if( time_float >= time_s and time_float <= time_e):


                if(len(cur) == 0):              #คือตัวแรกในเเต่ล่ะช่วงเวลา
                    cur.append(x)
                    keep_cur = deepcopy(cur)      

                else:

                    roo = 0

                    for j in range( len(cur) ):

                        if(cur[j].bus == x.bus):

                            # ระบุหมายเลขรถว่าหมายเลขเดียวกัน
                            roo = 1

                            #ห่าง น้อยกว่า 0.05 km หรือห่างแค่ 50 m คืออยู่ในจุดนั้นอยู่ คือรถมันยังอยู่ในจุด start จึงต้องเปลี่ยนเวลาของรถคันนั้น                        
                            if(dis_start <= 0.05 and x.speed != 0.0): 
                                keep_cur[j] = x

                            #เช็คว่าหมายเลขรถนั้นถึงจุดปลายทางหรือยัง ถ้าถึงเเล้วคิด avg เลย
                            elif(dis_destination <= 0.1):

                                #หาเวลาที่รถหมายเลขนั้นๆ ออกจากจุด start
                                time_previous = datetime.strptime((cur[j].timestamp), '%Y-%m-%dT%H:%M:%S.%f')
                                time_previous = time_previous.time()
                                
                                
#                                 time_previous = (cur[j].timestamp).split('T')
#                                 time_previous = time_previous[1].split(':')
#                                 time_previous = float(time_previous[0] + '.' + time_previous[1])

                                long_term = diff_times_in_seconds(time_previous , time_now)
                                Avg = Avg + long_term 
                                count = count + 1
                
#                                 print(long_term);
#                                 print ((time_now - time_previous))
#                                 print(diff_times_in_seconds(time_previous , time_now))
#                                 print("cur:    %d     %s  %f" %(cur[j].bus , cur[j].timestamp , cur[j].speed))
#                                 print(time)
#                                 print("x:    %d     %s  %f" %(x.bus , x.timestamp , x.speed))


                    if(roo == 0):
                        keep_cur.append(x)

                    cur = deepcopy(keep_cur)
        

        if(count == 0):
            Avg_all.append(0)

        else:
            Avg = Avg/count
            Avg_all.append(Avg)
            
        
    return (Avg_all)

#========================================
def avg(day_Mo):
    
    avg_Mo = []
    avg_All_Mo = []

    for i in day_Mo:
        keep = averageTime(i)
        avg_Mo.append(keep)

    for i in range(11):
        count = 0
        all_avg = 0
        for x in avg_Mo:
            if(x[i] != 0):
                count += 1

            all_avg = all_avg + x[i]

        all_avg = all_avg / count 
        avg_All_Mo.append(all_avg) 

    return avg_All_Mo  
```

```python
import math

def addDay(day, day_name, avg, place, data_all, route):

    time_star1 = ['07:00', '08:01', '09:31', '11:01', '12:31', '13:01', '14:31', '16:01', '17:31', '19:01', '20:31']
    time_end1 =  ['08:00', '09:30', '11:00', '12:30', '13:00', '14:30', '16:00', '17:30', '19:00', '20:30', '22:00']    
    
    for i in range (len(time_star1)):
        keep = []
        keep.append(day)
        keep.append(day_name)
        keep.append(time_star1[i])
        keep.append(time_end1[i])
        keep.append(math.ceil(avg[i]))
        keep.append(place)
        keep.append(route)
        
#         print(keep)
        
        data_all.append(keep)
        
    return data_all

count = 0
data_all = [] 
list_name_day = ['Thursday', 'Friday', 'Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday']
route = '3'

for i in range (len(list_day)):
    
    #หาว่าเป็นวันอะไร ==================
    keep_P = (list_day[i]).split('-')
    
    if(keep_P[1] == '3'):
        keep_P = int(keep_P[2]) + 28
    else:
        keep_P = int(keep_P[2])
    # ==========================

    data_all = addDay(list_day[i], list_name_day[(keep_P % 7)], data_day_avg_all[i], 'อาคารรัฐศาสตร์', data_all, route)
    
data_all
```
